```python
"""
=========================================
Asia Cup Cricket Data Analysis Project 🏏
=========================================

📌 PROJECT OVERVIEW
This script performs a comprehensive Exploratory Data Analysis (EDA) on 
historical Asia Cup cricket data, analyzing tournament trends, match results, 
and individual player performances across both ODI and T20I formats.

🎯 KEY OBJECTIVES
- Clean and structure tournament datasets (champions, match logs).
- Programmatically merge format-specific player performance statistics (ODI vs T20I).
- Standardize and clean categorical columns like match outcomes and toss details.
- Visualize player milestones and championship history using Seaborn & Matplotlib.

🛠️ DEPENDENCIES
Ensure you have the following libraries installed before running:
    $ pip install pandas numpy matplotlib seaborn

🚀 HOW TO RUN
Simply execute this script in your terminal or IDE:
    $ python asia_cup_analysis.py

Author: Abaid ur Rehman
Date: July 2026
"""

import pandas as pd
import numpy as np 
import seaborn as sp
import matplotlib.pyplot as pyt
```

```python

# --- Tournament & Match Level Data ---
df_cham = pd.read_csv('champion.csv')
df_mat = pd.read_csv('asiacup.csv')

# --- ODI Player Data ---
df_bat_odi = pd.read_csv('batsman data odi.csv')
df_bowl_odi = pd.read_csv('bowler data odi.csv')
df_keep_odi = pd.read_csv('wicketkeeper data odi.csv')

# --- T20I Player Data ---
df_bat_t20 = pd.read_csv('batsman data t20i.csv')
df_bowl_t20 = pd.read_csv('bowler data t20i.csv')
df_keep_t20 = pd.read_csv('wicketkeeper data t20i.csv')
```

```python
# Quick check of columns to plan your merges
print("Batsman ODI Columns:", df_bat_odi.columns.tolist())
print("Bowler ODI Columns:", df_bowl_odi.columns.tolist())
```

```python
player_dfs = [df_bat_odi, df_bowl_odi, df_keep_odi, df_bat_t20, df_bowl_t20, df_keep_t20]
```

```python
# 1. Take ODI and T20I batsmen data
# 2. Merge them ONLY on the 'Player Name' column
# 3. Use suffixes so we know which columns belong to which format
df_multi_format = pd.merge(
    df_bat_odi, 
    df_bat_t20, 
    on='Player Name', 
    how='inner', 
    suffixes=('_ODI', '_T20I')
)

# Now look at your beautifully combined custom table!
print(df_multi_format[['Player Name', 'Matches_T20I', 'Runs_T20I']].head())
```

```python
# 1. Take ODI and T20I bowler data
# 2. Merge them ONLY on the 'Player Name' column
# 3. Use suffixes so we know which columns belong to which format
# 1. Your merge is 100% correct!
df_multi_format = pd.merge(
    df_bowl_odi, 
    df_bowl_t20, 
    on='Player Name', 
    how='inner', 
    suffixes=('_ODI', '_T20I')
)

# 2. Fixed print statement using the new suffix names:
print(df_multi_format[['Player Name', 'Matches_ODI', 'Wickets_ODI', 'Matches_T20I', 'Wickets_T20I']].head())
```

```python
pyt.figure(figsize=(10, 6))
sp.set_theme(style="whitegrid")
sp.scatterplot(
    x='Economy Rate', 
    y='Wickets', 
    data=df_bowl_odi, 
    hue='Country', 
    s=100, 
    palette='Set2'
    
)

pyt.title('ODI Bowlers: Wickets Taken vs Economy Rate', fontsize=14, fontweight='bold', pad=15)
pyt.xlabel('Economy Rate (Lower is Better)', fontsize=12)
pyt.ylabel('Total Wickets Taken (Higher is Better)', fontsize=12)
pyt.tight_layout()
pyt.show()
```

```python
pyt.figure(figsize=(10, 6))
sp.set_theme(style="whitegrid")
sp.scatterplot(
    x='Strike Rate', 
    y='Runs', 
    data=df_bat_odi, 
    hue='Country', 
    s=120, 
    palette='Set1'
    
)

pyt.title('ODI Batsmen: Total Runs Scored vs Strike Rate', fontsize=14, fontweight='bold', pad=15)
pyt.xlabel('Strike Rate (Higher means faster scoring)', fontsize=12)
pyt.ylabel('Total Runs Scored', fontsize=12)
pyt.tight_layout()
pyt.show()
```

```python

# 1. Sort the data by 'Fifties' in descending order and get the top 5
top_fifties = df_bat_odi.sort_values(by='Fifties', ascending=False).head(10)

# 2. Setup the canvas size and design style
pyt.figure(figsize=(15, 5))
sp.set_theme(style="whitegrid")

# 3. Create the bar chart (X = Player Name, Y = Fifties)
sp.barplot(
    x='Player Name', 
    y='Fifties', 
    data=top_fifties, 
    hue='Country', 
    dodge=False, 
    palette='muted'
)

# 4. Add titles and axis labels
pyt.title('Players with the Most Fifties in Asia Cup History (ODI)', fontsize=14, fontweight='bold', pad=15)
pyt.xlabel('Player Name', fontsize=12)
pyt.ylabel('Number of 50s Scored', fontsize=12)

# 5. Render the chart cleanly
pyt.tight_layout()
pyt.show()
```

```python

team_wins = df_cham['Champion'].value_counts().reset_index()

# 2. Rename the columns so they are clean and easy to read
team_wins.columns = ['Team Name', 'Times Won']

# 3. Setup the canvas size and background grid
pyt.figure(figsize=(9, 5))
sp.set_theme(style="whitegrid")

# 4. Create the vertical bar chart (X = Team Name, Y = Times Won)
sp.barplot(
    x='Team Name', 
    y='Times Won', 
    data=team_wins, 
    hue='Team Name',    
    legend=False,
    palette='vlag'
)

# 5. Add clear titles and labels
pyt.title('Number of Asia Cup Titles Won by Each Team', fontsize=14, fontweight='bold', pad=15)
pyt.xlabel('Team Name', fontsize=12)
pyt.ylabel('How Many Times They Won', fontsize=12)

# 6. Show the final picture
pyt.tight_layout()
pyt.show()
```

```python

# 1. Filter rows where the team won the toss, and count how many times each team appears
toss_winners = df_mat[df_mat['Toss'] == 'Win']['Team'].value_counts().reset_index()

# 2. Rename the columns cleanly so they are easy to read
toss_winners.columns = ['Team Name', 'Tosses Won']

# 3. Setup the canvas size and background grid
pyt.figure(figsize=(9, 5))
sp.set_theme(style="whitegrid")

# 4. Create the vertical bar chart (X = Team Name, Y = Tosses Won)
sp.barplot(
    x='Team Name', 
    y='Tosses Won', 
    data=toss_winners, 
    hue='Team Name',     # Colors each bar by team name to keep it safe from warnings
    legend=False,        # Removes the extra duplicate box on the side
    palette='magma'      # Gives it a vibrant dark-to-light theme
)

# 5. Add clear titles and labels
pyt.title('Most Tosses Won by Teams in Asia Cup History', fontsize=14, fontweight='bold', pad=15)
pyt.xlabel('Team Name', fontsize=12)
pyt.ylabel('Number of Times Toss Won', fontsize=12)

# 6. Show the final picture cleanly
pyt.tight_layout()
pyt.show()
```
